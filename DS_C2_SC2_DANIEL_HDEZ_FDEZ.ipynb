Otra manera es conociendo el df e identificando los nombres de las vacunas de manera individual y aislar el dato por vacuna utilizada

```python
# Creación de la lista con nombres individuales de vacunas
lista_vacunas =['Johnson&Johnson',
                'Oxford/AstraZeneca', 
                'Pfizer/BioNTech', 
                'Sinopharm/Beijing', 
                'Sinovac', 
                'Sputnik V',
                'Sputnik Light', 
                'Moderna', 
                'CanSino', 
                'Covaxin', 
                'Novavax',
                'Abdala',
                'Soberana Plus',
                'Soberana02',
                'COVIran Barekat',
                'FAKHRAVAC',
                'Razi Cov Pars',
                'EpiVacCorona',
                'QazVac',
                'ZF2001',
                'Sinopharm/Wuhan',
                'Turkovac']

# df a iterar
df_vacunas = vaccinations['vaccines']
```

```python
# Creamos la función
def suma_vacunas(lista_vacunas, df_vacunas):

# Definimos un dic vacio para almacenar en item el nombre de la vacuna y en 
# value la suma de los registros en donde se encuentre el nombre de la vacuna
  total_vacunas = {}

# Con este for iteramos sobre los elementos de 'lista_vacunas'
  for nombre_vacuna in lista_vacunas:

# La var guardará las n veces que aparece el nombre de la vacuna en df_vacunas    
      aparece = 0 

# Usamos for anidado para iterar sobre el df_vacunas      
      for registro in df_vacunas:

# Evaluamos si el nombre de la vacuna en lista_vacunas aparece en df_vacunas
          if nombre_vacuna in registro:

# Ocupamos la función .count() de pd
              aparece += registro.count(nombre_vacuna)

# Asigmos en dic el item y su value
      total_vacunas[nombre_vacuna] = aparece

# Imprimimos por cada item
  for nombre_vacuna, aparece in total_vacunas.items():
      print(f"{nombre_vacuna}: {aparece}")
```

La intención de crear una función es para reducir código, ademas nos servirá
si se agrega una nueva vacuna a los registro o si estos cambian en la forma de la presentación de los datos.

```python
suma_vacunas(lista_unicos, vaccinations['vaccines'])
```

```python
suma_vacunas(lista_vacunas, df_vacunas)
```

Obtener la cantidad de vacunas aplicadas en todo el mundo y por cada país

En el df se visualiza la cantidad de vacunas diarias que se aplican en un pais, los datos son acumulados, se buscará el valor máximo en la columna 'total_vaccinations' que corresponda al país; posterior se hará la suma del dato obtenido por país para obtener el dato a nivel mundial



```python
# Se obtiene la lista de paises de 'country'
lista_paises = vaccinations['country'].unique()
```

```python
lista_paises
```

```python
# Creamos la función
def total_vacunas(lista_paises):
  total_vacunas_mundial = 0
  promedio_pais = 0
  print('***Según la columna "total_vaccinations" se muestra el total de vacunas que tenia cada país para su aplicación***\n')
  for nombre_pais in lista_paises:

# Usamos loc para ubicarnos en la columna 'country', y usamos .max() para
# obtener el valor máximo de la columna 'total
    cantidad_vacunas_pais = vaccinations.loc[vaccinations['country'] == nombre_pais, 'total_vaccinations'].max()
        
# Se sumará el número máximo de vacunas por país
    total_vacunas_mundial += cantidad_vacunas_pais
    
    # Se tiene duda sobre el dato solicitado en este apartado
    # --- promedio_pais = (cantidad_vacunas_pais / total_vacunas_mundial) ---
    
    print(f'En {nombre_pais} el total de vacunas aplicadas fueron {cantidad_vacunas_pais:,}')
    
  print(f'\nEl total de vacunas aplicadas en el mundo fue de {total_vacunas_mundial:,}')
 
```

```python
total_vacunas(lista_paises)
```

Crear un df nuevo llamado "conCantidad" con los datos del df original y adicionar una columna "canVac" con la cantidad de vacunas usadas cada día según la columna "vaccines" y separar por el carácter "/".

Nota.- Para este ejercicio use como separador el carácter "," ya que es este el que sirve de separador entre los nombres de las vacunas.

```python
# Copiamos el df con .copy()
conCantidad = vaccinations.copy()
```

```python
# Verificamos se haya creado OK
conCantidad
```

```python
# Con .str.split() llamamos a la función que separa según el carácter indicado y usamos .apply() para aplicar la función len en la misma línea de código
conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').apply(len)
```
